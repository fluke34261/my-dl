```python
from glob import glob
from tqdm import tqdm
import pandas as pd
import shutil

file_marksheet = 'picai_labels/clinical_information/marksheet.csv'
root_dir = '/media/mpmri/mpmri/nnUNet_raw_data/Task2201_picai_baseline/'
```

```python
marksheet = pd.read_csv(file_marksheet)
logic = marksheet['lesion_ISUP'].str.contains('3') | \
        marksheet['lesion_ISUP'].str.contains('4') | \
        marksheet['lesion_ISUP'].str.contains('5')
marksheet = marksheet[logic]
```

```python
marksheet
```

```python
for mi, m in tqdm(marksheet.iterrows(), total=len(marksheet)):
    file_image = f"{root_dir}/imagesTr/{m['patient_id']}_{m['study_id']}*"
    file_image = glob(file_image)
    file_label = f"{root_dir}/labelsTr/{m['patient_id']}_{m['study_id']}*"
    file_label = glob(file_label)
    for k in file_image:
        shutil.copyfile(k, f"cases/images/{k.split('/')[-1]}")
    for k in file_label:
        shutil.copyfile(k, f"cases/labels/{k.split('/')[-1]}")
```
