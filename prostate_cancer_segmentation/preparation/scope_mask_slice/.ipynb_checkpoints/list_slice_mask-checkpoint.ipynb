```python
import nibabel as nib
import numpy as np
import glob as glob
import shutil
```

```python
extract_label_path = '../../extract_data/each_labels/'
mask_label_path = './mask_labels/'

extract_image_path = '../../extract_data/each_images/'
mask_image_path = './mask_images/'
```

```python
nifti_lists = sorted(glob.glob(extract_label_path + '*'))
```

```python
nifti_lists
```

```python
def readNifti(nifti_file):
    img = nib.load(nifti_file)
    return img.get_fdata()
```

```python
def getImageSequenceFromLabel(label_name):
    t2w = label_name[0:13] + "_0000" + label_name[13:]
    adc = label_name[0:13] + "_0001" + label_name[13:]
    dwi = label_name[0:13] + "_0002" + label_name[13:]
    
    return [t2w, adc, dwi]
```

```python
def isMask(img_np_array):
    unique_result = np.unique(img_np_array)
    return unique_result.size > 1
```

```python
def copyMaskLabelAndImage(nifti_file):
    nifti_target_file = nifti_file.replace(extract_label_path, "")
    
    # copy label
    shutil.copy(
            extract_label_path + nifti_target_file, 
            mask_label_path + nifti_target_file
    )
    
    t2w, adc, dwi = getImageSequenceFromLabel(nifti_target_file)
    # copy image
#     shutil.copy(
#             extract_image_path + t2w, 
#             mask_image_path + t2w
#     )
    
    shutil.copy(
            extract_image_path + adc, 
            mask_image_path + adc
    )
    
    shutil.copy(
            extract_image_path + dwi, 
            mask_image_path + dwi
    )
```

```python
for nifti in nifti_lists:
    img = readNifti(nifti)
    if isMask(img):
        copyMaskLabelAndImage(nifti)
```
